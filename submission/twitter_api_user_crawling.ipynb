```python
# drop useless columns
users = panda_tweets[[
    'userid'
]]
#  get unique users
users = users.drop_duplicates(subset=['userid'])
print(users.shape)

users = np.array(users)
```

```python
# CRAWL THE FOLLOWERS OF A USER
count = 0
followers_list = []

for user in users:
    
    count = count + 1
    # comment this 'if statement' if you want to crawl all the users
#     if (count == 3):
#         break
    print(count)
    
    url =  "https://api.twitter.com/1.1/followers/list.json?cursor=-1&user_id=" + str(user[0]) + "&skip_status=true&include_user_entities=false"
    headers = {'Authorization': 'Bearer AAAAAAAAAAAAAAAAAAAAALy7aAEAAAAAlq8iM%2BNos2N1emkKZ3wn%2B0rs%2Bd0%3DpUL4vmOfD9OXE0yUMIi6If7ftHVUjdZ5gq55yKmIyXcGLEbgUK'}
    followers = requests.get(url, headers=headers)
    followers = followers.json()
    
    # handle a possible error
    if (followers.get('errors') is not None or followers.get('error') is not None):
        print(followers)
        print('Error for user:' + str(user[0]))
        continue
    
    followers = followers['users']
    followers = [follower['screen_name'] for follower in followers]
    followers_list.append([user[0], followers])
    
    
    # only one request per MINUTE  can be made
    time.sleep(60.1)

pd.DataFrame(followers_list, columns=['user', 'followers']).to_csv('followers.csv', index=False)    
```
